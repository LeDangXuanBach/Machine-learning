# Task 2.1. dataset1.csv


```python
#code
data1 = pd.read_csv('dataset1.csv')
X_train = data1[['x']]
y_train = data1[['y']]
X_train
regressor = LinearRegression()
regressor.fit(X_train, y_train)
# b
print(regressor.intercept_)
# a
print(regressor.coef_)

# y = 0.4 + 0.8*x
y_pred = regressor.predict(X_train)
y_pred

```

# Task 2.2. dataset2.csv


```python
#code
data2 = pd.read_csv("dataset2.csv")
print("Cơ sở dữ liệu")
print(data2)
# Index(['id', 'x1', 'x2', 'x3', 'y'], dtype='object')
# chia dữ liệu thành các tập train(0.7) và test(0.3)
X = data2[['x1', 'x2', 'x3']]
y = data2[['y']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Tạo một đối tượng LinearRegression
reg = LinearRegression()

# Huấn luyện mô hình trên tập huấn luyện
reg.fit(X_train, y_train)

print("\n\nDự đoán giá trị trên tập kiểm tra")
y_pred = reg.predict(X_test)
print(y_pred)
print("\n\nĐánh giá độ chính xác của mô hình")
score = reg.score(X_test, y_test)
print(score)

print("\n\nCác hệ số")
print(reg.intercept_)
print(reg.coef_)

```

# Task 2.3. Apply LinearRegression to the following dataset:
*   Red wine quality: https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv . The dataset includes info about the chemical properties of different types of wine and how they relate to overall quality. 



```python
#code
data3 = pd.read_csv('winequality-red.csv', sep = ';')
print("Cơ sở dữ liệu đánh giá rượu vang đỏ\n")
print(data3.head(len(data3)))
#Index(['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality'], dtype='object')
X3 = data3[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
y3 = data3[['quality']]
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.3, random_state=43)

# Tạo một đối tượng LinearRegression
reg3 = LinearRegression()

# Huấn luyện mô hình trên tập huấn luyện
reg3.fit(X_train3, y_train3)

y_pred3 = reg3.predict(X_test3)
print("\n\nDự đoán giá trị trên tập kiểm tra\n", y_pred3)

print("\n\nĐánh giá độ chính xác của mô hình")
score3 = reg3.score(X_test3, y_test3)
print(score3)

print("\n\nCác hệ số")
print(reg3.intercept_)
print(reg3.coef_)
```

# Task 2.4. Apply LinearRegression to the following dataset:
*   Energy efficiency: https://archive.ics.uci.edu/ml/datasets/Energy+efficiency.




```python
#code
data4 = pd.read_excel('ENB2012_data.xlsx')
print("Cơ sở dữ liệu")
print(data4)
#Index(['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8', 'Y1', 'Y2'], dtype='object')
X4 = data4[['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8']]
y41 = data4[['Y1']]
y42 = data4[['Y2']]
X_train4, X_test4, y_train41, y_test41, y_train42, y_test42 = train_test_split(X4, y41, y42, test_size=0.3, random_state=44)

# Mô hình 4.1
print("\n\nMô Hình 4.1\n\n")
# Tạo một đối tượng LinearRegression
reg41 = LinearRegression()

# Huấn luyện mô hình trên tập huấn luyện
reg41.fit(X_train4, y_train41)

y_pred41 = reg41.predict(X_test4)
print("\n\nDự đoán giá trị trên tập kiểm tra của mô hình 4.1\n", y_pred41)

print("\n\nĐánh giá độ chính xác của mô hình 4.1")
score41 = reg41.score(X_test4, y_test41)
print(score41)

print("\n\nCác hệ số")
print(reg41.intercept_)
print(reg41.coef_)

# Mô hình 4.2
print("\n\nMô Hình 4.2\n\n")
# Tạo một đối tượng LinearRegression
reg42 = LinearRegression()

# Huấn luyện mô hình trên tập huấn luyện
reg42.fit(X_train4, y_train42)

y_pred42 = reg42.predict(X_test4)
print("\n\nDự đoán giá trị trên tập kiểm tra của mô hình 4.2\n", y_pred42)

print("\n\nĐánh giá độ chính xác của mô hình 4.2")
score42 = reg42.score(X_test4, y_test42)
print(score42)

print("\n\nCác hệ số")
print(reg42.intercept_)
print(reg42.coef_)
```
